```python
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import mplcursors
%matplotlib inline

style.use('ggplot')
```

```python
def plotter(xl, sheet, type):
    data = pd.read_excel(xl,sheet)
    plt.figure(figsize=(15,5))
    plt.scatter(data['COMPONENT-NO.'], data['% ACCURACY'])
    plt.xlabel('Components')
    plt.ylabel('Accuracy %')
    plt.title(type+'classifier performance on GLCM data')
    mplcursors.cursor(hover=True)
    plt.show()
    #plt.savefig(path+'{}.png'.format(serial))
```

```python
df = pd.ExcelFile("E:\\THESIS\\ADNI_data\\ADNI1_Annual_2_Yr_3T_306_WORK\\LogRegClassifier\\all_GLCM_results.xlsx")
#dtree_sheet = pd.read_excel(df,'dtree_unique_max')
plotter(df,'dtree_unique_max','Decision tree')
```
